```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
```

```python
data_path = 'data/hindi_hatespeech.tsv'

save_train_path = 'data/hindi_hatespeech_cv_train.csv'
save_val_path = 'data/hindi_hatespeech_cv_val.csv'

# data_path = 'Task_2/hindi_bengali/save/bengali_hatespeech_sample_train_preprocessed.csv'

# save_train_path = 'Task_2/hindi_bengali/save/bengali_hatespeech_sample_train_preprocessed_cv_train.csv'
# save_val_path = 'Task_2/hindi_bengali/save/bengali_hatespeech_sample_train_preprocessed_cv_val.csv'
```

```python
data = pd.read_csv(data_path, sep='\t')
```

```python
data
```

```python
data_train, data_val = train_test_split(data, train_size=0.8, stratify=data['task_1'])
```

```python
data_train.to_csv(save_train_path, index=False)
data_val.to_csv(save_val_path, index=False)
```
