## Import libraries

```python
import os
import json

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go

sns.set()
```

# Figure 1

## Load data

```python
hindi = pd.read_csv('./data/hindi_hatespeech.tsv', sep='\t')
sample_bengali = pd.read_csv('./Task_2/hindi_bengali/save/bengali_hatespeech_sample_train.csv')
bengali = pd.read_csv('./data/bengali_hatespeech.csv')

# convert hindi's columns to match that of bengali
hindi['hate'] = (hindi['task_1'] == 'HOF').astype(int)
hindi.rename(columns={'text':'sentence'}, inplace=True)
```

## Auxiliary functions

```python
# remove punctuations
import string
punctuation = string.punctuation[:2] + string.punctuation[3:]
translator = str.maketrans(punctuation, ' '*len(punctuation))
def remove_puncs(s):
    s = s.translate(translator)
    return s
```

## Plot statistics

```python
fig = plt.figure(figsize=(15, 8))

# number of texts
ax1 = plt.subplot2grid((2, 3), (0, 0), colspan=1)

corpus_sizes = np.array([len(hindi), len(sample_bengali), len(bengali)])
ax1 = sns.barplot(ax=ax1, orient='v', y=corpus_sizes, x=[0, 1, 2])
ax1.set_xticklabels(['Hindi', 'Sample Bengali', 'Bengali'])
ax1.text(0, corpus_sizes[0], corpus_sizes[0], ha='center')
ax1.text(1, corpus_sizes[1], corpus_sizes[1], ha='center')
ax1.text(2, corpus_sizes[2], corpus_sizes[2], ha='center')
ax1.set_title('a. Number of texts', loc='left')

# average tokens per text
ax2 = plt.subplot2grid((2, 3), (0, 1), colspan=1)

hindi_sentences = [remove_puncs(sentence) for sentence in hindi['sentence']]
sample_bengali_sentences = [remove_puncs(sentence) for sentence in sample_bengali['sentence']]
bengali_sentences = [remove_puncs(sentence) for sentence in bengali['sentence']]

hindi_flattened = [w for sentence in hindi_sentences for w in sentence.split()]
sample_bengali_flattened = [w for sentence in sample_bengali_sentences for w in sentence.split()]
bengali_flattened = [w for sentence in bengali_sentences for w in sentence.split()]

tokens = np.array([len(hindi_flattened), len(sample_bengali_flattened), len(bengali_flattened)])
tokens_per_text = tokens / corpus_sizes

ax2 = sns.barplot(ax=ax2, orient='v', y=tokens_per_text, x=[0, 1, 2])
ax2.set_xticklabels(['Hindi', 'Sample Bengali', 'Bengali'])
ax2.text(0, tokens_per_text[0], np.round(tokens_per_text[0], 2), ha='center')
ax2.text(1, tokens_per_text[1], np.round(tokens_per_text[1], 2), ha='center')
ax2.text(2, tokens_per_text[2], np.round(tokens_per_text[2], 2), ha='center')
ax2.set_title('b. Average tokens per text', loc='left')

# vocab size
ax3 = plt.subplot2grid((2, 3), (0, 2), colspan=1)

hindi_vocab = sorted(list(set(hindi_flattened)))
sample_bengali_vocab = sorted(list(set(sample_bengali_flattened)))
bengali_vocab = sorted(list(set(bengali_flattened)))
vocab_sizes = np.array([len(hindi_vocab), len(sample_bengali_vocab), len(bengali_vocab)])

ax3 = sns.barplot(ax=ax3, orient='v', y=vocab_sizes, x=[0, 1, 2])
ax3.set_xticklabels(['Hindi', 'Sample Bengali', 'Bengali'])
ax3.text(0, vocab_sizes[0], np.round(vocab_sizes[0], 2), ha='center')
ax3.text(1, vocab_sizes[1], np.round(vocab_sizes[1], 2), ha='center')
ax3.text(2, vocab_sizes[2], np.round(vocab_sizes[2], 2), ha='center')
ax3.set_title('c. Vocab size', loc='left')

# Average occurrences per token
ax4 = plt.subplot2grid((2, 3), (1, 0), colspan=1)

hindi_occurrences_per_token = len(hindi_flattened) / len(hindi_vocab)
sample_bengali_occurrences_per_token = len(sample_bengali_flattened) / len(sample_bengali_vocab)
bengali_occurrences_per_token = len(bengali_flattened) / len(bengali_vocab)
list_occurrences_per_token = [hindi_occurrences_per_token, 
                              sample_bengali_occurrences_per_token, 
                              bengali_occurrences_per_token]

sns.barplot(ax=ax4, orient='v', y=list_occurrences_per_token, x=[0, 1, 2])
ax4.set_xticklabels(['Hindi', 'Sample Bengali', 'Bengali'])
ax4.text(0, list_occurrences_per_token[0], f'{np.round(list_occurrences_per_token[0], 2)}', ha='center')
ax4.text(1, list_occurrences_per_token[1], f'{np.round(list_occurrences_per_token[1], 2)}', ha='center')
ax4.text(2, list_occurrences_per_token[2], f'{np.round(list_occurrences_per_token[2], 2)}', ha='center')
# ax4.set_ylabel('proportion (%)')
ax4.set_title('d. Average no. occurrences per distinct token', loc='left')

# Label distribution
ax5 = plt.subplot2grid((2, 3), (1, 1), colspan=1)

hindi_hof_prop = hindi['hate'].mean() * 100
sample_bengali_hof_prop = sample_bengali['hate'].mean() * 100
bengali_hof_prop = bengali['hate'].mean() * 100
hof_props = [hindi_hof_prop, sample_bengali_hof_prop, bengali_hof_prop]

sns.barplot(ax=ax5, orient='v', y=hof_props, x=[0, 1, 2])
ax5.set_xticklabels(['Hindi', 'Sample Bengali', 'Bengali'])
ax5.text(0, hof_props[0], f'{np.round(hof_props[0], 2)}%', ha='center')
ax5.text(1, hof_props[1], f'{np.round(hof_props[1], 2)}%', ha='center')
ax5.text(2, hof_props[2], f'{np.round(hof_props[2], 2)}%', ha='center')
ax5.set_ylabel('proportion (%)')
ax5.set_title('e. Hate/Offensive proportions', loc='left')

# save and show
plt.savefig('figures/figure_1.png')
plt.show()
```

# Figure 3

```python
# function to load performance result
def get_perf(task=2, model_type='hindi_hindi', metric='accuracy', fix='no_fix', take_mean=False):
    data_path = f'Task_{task}/{model_type}/save/test_result/{metric}_{fix}.txt'
    if os.path.isfile(data_path):
        with open(data_path, 'r') as f:
            data = json.load(f)
        if take_mean:
            data = np.mean(data, axis=0)
    else:
        data = None
    return data
```

```python
fig = go.Figure(layout=dict(height=500, width=1000))


t2_hinhin_fix = get_perf(2, 'hindi_hindi', 'accuracy', 'no_fix', take_mean=True)
fig.add_trace(
    go.Scatter(
        x=list(range(1, 16)),
        y=t2_hinhin_fix,
        mode='lines',
        line=dict(color='blue', dash='dash'),
        name='Baseline Hindi-Hindi',
))

t2_hinben_fix = get_perf(2, 'hindi_bengali', 'accuracy', 'no_fix', take_mean=True)
fig.add_trace(
    go.Scatter(
        x=list(range(1, 16)),
        y=t2_hinben_fix,
        mode='lines',
        line=dict(color='green', dash='dash'),
        name='Baseline Hindi-Bengali',
))

t2_benben_fix = get_perf(2, 'bengali_bengali', 'accuracy', 'no_fix', take_mean=True)
fig.add_trace(
    go.Scatter(
        x=list(range(1, 16)),
        y=t2_benben_fix,
        mode='lines',
        line=dict(color='orange', dash='dash'),
        name='Baseline Bengali-Bengali',
))

t3_hinhin_fix = get_perf(3, 'hindi_hindi', 'accuracy', 'no_fix', take_mean=True)
fig.add_trace(
    go.Scatter(
        x=list(range(1, 16)),
        y=t3_hinhin_fix,
        mode='lines',
        line=dict(color='blue'),
        name='Enhanced Hindi-Hindi',
))

t3_hinben_fix = get_perf(3, 'hindi_bengali', 'accuracy', 'no_fix', take_mean=True)
fig.add_trace(
    go.Scatter(
        x=list(range(1, 16)),
        y=t3_hinben_fix,
        mode='lines',
        line=dict(color='green'),
        name='Enhanced Hindi-Bengali',
))

t3_benben_fix = get_perf(3, 'bengali_bengali', 'accuracy', 'no_fix', take_mean=True)
fig.add_trace(
    go.Scatter(
        x=list(range(1, 16)),
        y=t3_benben_fix,
        mode='lines',
        line=dict(color='orange'),
        name='Enhanced Bengali-Bengali',
))

fig.update_layout(
    xaxis_title='Epoch',
    yaxis_title='Accuracy',
    title='6 models with no fixed layers, Accuracy'
)

fig.show()
```

```python
print('Hindi-Hindi rises:', (t3_hinhin_fix[10:] - t2_hinhin_fix[10:]).mean())
print('Hindi-Bengali rises:', (t3_hinben_fix[10:] - t2_hinben_fix[10:]).mean())
print('Bengali-Bengali rises:', (t3_benben_fix[10:] - t2_benben_fix[10:]).mean())
```

# Figure 4

```python
def plot_results(list_test_macro_f1):
    # macro-f1 plot
    mean_mac = [np.mean([l[i] for l in list_test_macro_f1 if len(l) > i]) for i in range(max(len(li) for li in list_test_macro_f1))]

    fig = go.Figure(layout=dict(height=500, width=1000))
    for i, mac_curve in enumerate(list_test_macro_f1):
        fig.add_trace(go.Scatter(
            x=list(range(1, len(mac_curve) + 1)), 
            y=mac_curve,
            mode='lines',
            name=f'model {i+1}',
            opacity=0.3,
        ))
    fig.add_trace(go.Scatter(
        x=list(range(1, len(mean_mac) + 1)),
        y=mean_mac,
        mode='lines+markers',
        line_color='blue',
        name='mean',
    ))

    fig.update_layout(
        title='Enhanced Hindi-Hindi model, Macro-f1',
        xaxis_title='Epoch',
        yaxis_title='Macro-f1'
    )

    fig.show()
```

```python
t3_hinhin_mf1_no_fix = get_perf(3, 'hindi_hindi', 'macro_f1', 'no_fix', take_mean=False)
plot_results(t3_hinhin_mf1_no_fix)
```
