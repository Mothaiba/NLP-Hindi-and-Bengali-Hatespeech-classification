```python
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 25})
import numpy as np
```

```python
max_len = 128
d = 512

def get_pe(p, i):
    k = i // 2
    if i % 2 == 0:
        return np.sin(p/(10000**(2*k/d)))
    else:
        return np.cos(p/(10000**(2*k/d)))

pe = np.zeros((d, max_len))
for p in range(max_len):
    for i in range(d):
        pe[i, p] = get_pe(p, i)
```

```python
fig, ax = plt.subplots(figsize=(16, 10))
ax.imshow(pe.T[:,:512])
ax.set_xlabel('dimension')
ax.set_ylabel('position')
```

```python
import seaborn as sns
```

```python
def get_cosine_similarity(a, b):
    return np.dot(a, b)/(np.linalg.norm(a)*np.linalg.norm(b))

l = 128

pe_list = []
for p in range(l):
    a = np.zeros((d))
    for z in range(d):
        a[z] = get_pe(p, z)
    pe_list.append(a)

dmap = np.zeros((l, l))
for i in range(l):
    for j in range(l):
        dmap[i, j] = get_cosine_similarity(pe_list[i], pe_list[j])

plt.rcParams.update({'font.size': 15})
plt.subplots(figsize=(10, 8))
sns.heatmap(dmap)
plt.title('Cosine similarity of Positional Encodings')
plt.xlabel('position')
plt.ylabel('position')
```
